```python
import pandas as pd
register20230717_df = pd.read_csv("/Users/leeliang/Desktop/dashboard_registration/registerdata20230717.csv")
register20230717_df
```

```python
# putting lastname in firstname column
for i in range(register20230717_df.shape[0]):
    if register20230717_df["lastname"].iloc[i] == '""':
        new_firstname = register20230717_df["firstname"].iloc[i][1:]
        new_lastname = register20230717_df["firstname"].iloc[i][0]
        register20230717_df["lastname"].iloc[i] = new_lastname
        register20230717_df["firstname"].iloc[i] = new_firstname

register20230717_df.head(20)
```

## email/domain

```python
email_list = register20230717_df["email"].values
domain_name_list = [i[i.index('@') + 1 : ] for i in email_list]
domain_name_unqiue_list = list(set(domain_name_list))
domain_name_unqiue_list
```

```python

domain_list = ["edu.tw", "gmail", "qq", "yahoo", "hotmail"]
each_domain_count_list = [len([i for i in domain_name_list if d in i]) for d in domain_list]
each_domain_count_list
```

```python
data = {'domain': domain_list + ["other"],
        'count': each_domain_count_list + [len(domain_name_list) - sum(each_domain_count_list)]}

domain_count_df = pd.DataFrame(data)
domain_count_df
```

```python
domain_count_df['domain'][domain_count_df['count'] == domain_count_df['count'].max()].values[0]
```
