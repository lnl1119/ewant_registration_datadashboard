```python
import pandas as pd
login_2016_2022_df = pd.read_csv("/Users/leeliang/Desktop/dashboard_registration/login_2016-2022.csv")
login_2016_2022_df
```

```python
login_melt_df = login_2016_2022_df.melt(id_vars=["date"])
date_done_df = pd.to_datetime(login_melt_df["variable"] + "-" + login_melt_df["date"])
date_login_df = pd.concat([date_done_df.rename("date"), login_melt_df["value"]], axis = 1)
date_login_df
```

```python
date_login_df.to_csv("/Users/leeliang/Desktop/dashboard_registration/login_20162022_done.csv")
```
